**Pyspark Handling Missing Values**

*   Dropping Columns & Rows
*   Various Parameter in dropping functionalities
*   Handling Missing values by mean, Median Mode


```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Practise').getOrCreate()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df_pyspark = spark.read.csv('test2.csv', header=True, inferSchema=True)
df_pyspark.show()
```

```python
df_pyspark.drop('age').show()
```

```python
df_pyspark.na.drop().show()
```

```python
df_pyspark.na.drop(how="any").show()
df_pyspark.na.drop(how="any",thresh=1).show()
df_pyspark.na.drop(how="any",subset=['experinece']).show()
```

```python
df_pyspark.na.fill('Missing Values').show()
df_pyspark.na.fill('Missing Values',['experinece','age']).show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=['age', 'experinece', 'salary'],
    outputCols=["{}_imputed".format(c) for c in ['age', 'Experience', 'Salary']]
    ).setStrategy("median")
```

```python
imputer.fit(df_pyspark).transform(df_pyspark).show()
```

**Pyspark Dataframe**

1.   Filter Operation
2.   & | ==

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Dataframe').getOrCreate()
```

```python
df_pyspark.filter("salary<=20000").show()
df_pyspark.filter("salary<=20000").select(['name ','age']).show()
df_pyspark.filter((df_pyspark['salary']<=40000)&
                  (df_pyspark['salary']>=20000)).show()
df_pyspark.filter((df_pyspark['salary']<=40000)|
                  (df_pyspark['salary']>=20000)).show()
```

```python
df_pyspark.filter(~(df_pyspark['salary']<=20000)).show()
```

**Pyspark Group by and aggregate functions**

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Agg').getOrCreate()
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df_pyspark = spark.read.csv('test3.csv', header=True, inferSchema=True)
df_pyspark.show()
```

```python
df_pyspark.printSchema()
```

```python
df_pyspark.groupBy('department')
df_pyspark.groupBy('name')
```

```python
df_pyspark.groupBy('department').sum().show()
df_pyspark.groupBy('department').mean().show()
df_pyspark.groupBy('department').count().show()
df_pyspark.groupBy('department').max().show()
df_pyspark.groupBy('department').min().show()
df_pyspark.groupBy('department').avg().show()
df_pyspark.agg({'Salary':'sum'}).show()


###df_pyspark.groupBy('name').sum().show()
```
